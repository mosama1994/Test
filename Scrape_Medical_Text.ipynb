```python
def create_text(text_list):
  full_text = ''

  header_dict = {
      'h1': 0,
      'h2': 0,
      'h3': 0,
      'h4': 0,
      'h5': 0,
      'h6': 0
  }

  for i in text_list:
    if 'h' in i.name:
      header_dict[i.name] += 1

  list_header = list(header_dict.values())

  for i in range(len(list_header)):
    if list_header[i] > 0:
      top = i + 1
      break

  top_heading = 'h' + str(top)
  lower_range = list(range(top + 1,7))
  lower_heading = ['h' + str(i) for i in lower_range]

  list_li = []

  for i in text_list:
    if 'li' in i.name:
      list_li.append(text_list.index(i))

  # print(list_li)

  list_li_split = []
  count = 1
  new_list = []
  for i in range(1,len(list_li)):
    if i == 1:
      new_list.append(count)
      count = count + 1
    if list_li[i] == (list_li[i-1] + 1):
      new_list.append(count)
      count = count + 1
      if i == len(list_li) - 1:
        list_li_split.append(new_list)
    else:
      list_li_split.append(new_list)
      new_list = []
      count = 1
      new_list.append(count)
      count = count + 1
      if i == len(list_li) - 1:
        list_li_split.append(new_list)

  count_x = 0 # outer list index
  count_y = 0 # inner list index

  for n in text_list:
    if n.name == 'table':
      text = preprocess_table(n)

    else:
      text = n.text.strip()
      if text != '':
        # print(top_heading, n.name)
        if top_heading in n.name:
          text = "\n\n<header>" + text + "</header>\n"

        if n.name in lower_heading:
          text = text + ":"

        if 'li' in n.name:
          # print(count_x, count_y)
          number = list_li_split[count_x][count_y]
          text = " " + str(number) + '. ' + text
          if len(list_li_split[count_x]) == count_y + 1:
            count_x += 1
            count_y = 0
          else:
            count_y += 1

    full_text = full_text + "\n" + text
  full_text = unicodedata.normalize("NFKD", full_text)

  return full_text
```

```python
# Functions for table text preprocessing

def preprocess_table(temp):
  alphabets = list(string.ascii_lowercase)
  header_list = []
  row_list = []
  for i in temp.find_all('tr'):
    header = i.find_all('th')
    if header:
      for j in header:
        texts = j.get_text(separator=' ',strip=True)
        header_list.append(texts)

    row = i.find_all('td')
    row_text = []
    if row:
      # print(row)
      for j in row:
        if j.find_all('li'):
          full_list_text = ' '
          list_li_tag = j.find_all('li')
          for d in list_li_tag:
            index_list = list_li_tag.index(d)
            if d.get_text(strip=True) != '':
              list_text = d.get_text(separator=' ',strip=True)
              full_list_text = full_list_text + alphabets[index_list] + ') ' + list_text + ', '
          full_list_text = full_list_text[:-2]
          row_text.append(full_list_text)
        else:
          if j.get_text(strip=True) != '':
            row_text.append(j.get_text(separator=' ',strip=True))
    if row_text:
      row_list.append(row_text)

  # print(header_list)
  # print(row_list)
  table_text = '\n'
  if header_list:
    for i in row_list:
      text = ''
      index_row = row_list.index(i) + 1
      for j in range(len(i)):
        if header_list[j].strip() == '':
          # print('here')
          if j == 0:
            text = text + ' ' + str(index_row) + '. ' + i[j].strip() + '\n'
          else:
            text = text + '    ' + i[j].strip() + '\n'
        else:
          if j == 0:
            text = text + ' ' + str(index_row) + '. ' + header_list[j].strip() + ': ' + i[j].strip() + '\n'
          else:
            text = text + '    ' + header_list[j].strip() + ': ' + i[j].strip() + '\n'
      table_text = table_text + text
  else:
    for i in row_list:
      text = ''
      for j in range(len(i)):
        if j == 0:
          text = text + ' > ' + i[j].strip() + '\n'
        else:
          text = text + '   ' + i[j].strip() + '\n'
      table_text = table_text + text
  return table_text
```

```python
# Remove unwanted information like Products and Services & More Information

def remove_info(info, all_p):
  flag = 0
  present = 0
  ind = -1
  start_1 = -1
  end_1 = -1

  for i in range(len(all_p)):
    text = all_p[i].text.strip()

    if 'h' in all_p[i].name and info in text:
      start_1 = i
      flag = 1
      present = 1
      continue

    if 'h' in all_p[i].name and flag == 1:
      end_1 = i
      flag = 0

  if present == 1:
    all_p_index = list(range(len(all_p)))
    if end_1 == -1:
      remove_index = list(range(start_1, len(all_p)))
    else:
      remove_index = list(range(start_1, end_1))
    all_p = [all_p[i] for i in all_p_index if i not in remove_index]

  return all_p
```

```python
def create_text(text_list):
  full_text = ''
  for n in text_list:
    text = n.text.strip()

    if 'h2' in n.name:
      text = "\n\n<header>" + text + "</header>\n"

    if n.name in ['h3','h4','h5','h6']:
      text = text + ":"

    if 'li' in n.name:
      text = "  > " + text

    full_text = full_text + "\n" + text
    full_text = unicodedata.normalize("NFKD", full_text)

  return full_text
```

```python
def create_text(text_list):
  full_text = ''

  header_dict = {
      'h1': 0,
      'h2': 0,
      'h3': 0,
      'h4': 0,
      'h5': 0,
      'h6': 0
  }

  for i in text_list:
    if 'h' in i.name:
      header_dict[i.name] += 1

  list_header = list(header_dict.values())

  for i in range(len(list_header)):
    if list_header[i] > 0:
      top = i + 1
      break

  top_heading = 'h' + str(top)
  lower_range = list(range(top + 1,7))
  lower_heading = ['h' + str(i) for i in lower_range]

  list_li = []

  for i in text_list:
    if 'li' in i.name:
      list_li.append(text_list.index(i))

  print(list_li)

  list_li_split = []
  count = 1
  new_list = []
  for i in range(1,len(list_li)):
    if i == 1:
      new_list.append(count)
      count = count + 1
    if list_li[i] == (list_li[i-1] + 1):
      new_list.append(count)
      count = count + 1
      if i == len(list_li) - 1:
        list_li_split.append(new_list)
    else:
      list_li_split.append(new_list)
      new_list = []
      count = 1
      new_list.append(count)
      count = count + 1
      if i == len(list_li) - 1:
        list_li_split.append(new_list)

  count_x = 0 # outer list index
  count_y = 0 # inner list index

  for n in text_list:
    if n.name == 'table':
      text = preprocess_table(n)

    else:
      text = n.text.strip()
      if text != '':
        print(top_heading, n.name)
        if top_heading in n.name:
          text = "\n\n<header>" + text + "</header>\n"

        if n.name in lower_heading:
          text = text + ":"

        if 'li' in n.name:
          print(count_x, count_y)
          number = list_li_split[count_x][count_y]
          text = " " + str(number) + '. ' + text
          if len(list_li_split[count_x]) == count_y + 1:
            count_x += 1
            count_y = 0
          else:
            count_y += 1

    full_text = full_text + "\n" + text
  full_text = unicodedata.normalize("NFKD", full_text)

  return full_text
```

```python
# Functions for table text preprocessing

def preprocess_table(temp):
  alphabets = list(string.ascii_lowercase)
  header_list = []
  row_list = []
  for i in temp.find_all('tr'):
    header = i.find_all('th')
    if header:
      for j in header:
        texts = j.get_text(separator=' ',strip=True)
        header_list.append(texts)

    row = i.find_all('td')
    row_text = []
    if row:
      print(row)
      for j in row:
        if j.find_all('li'):
          full_list_text = ' '
          list_li_tag = j.find_all('li')
          for d in list_li_tag:
            index_list = list_li_tag.index(d)
            if d.get_text(strip=True) != '':
              list_text = d.get_text(separator=' ',strip=True)
              full_list_text = full_list_text + alphabets[index_list] + ') ' + list_text + ', '
          full_list_text = full_list_text[:-2]
          row_text.append(full_list_text)
        else:
          if j.get_text(strip=True) != '':
            row_text.append(j.get_text(separator=' ',strip=True))
    if row_text:
      row_list.append(row_text)

  print(header_list)
  print(row_list)
  table_text = '\n'
  if header_list:
    for i in row_list:
      text = ''
      index_row = row_list.index(i) + 1
      for j in range(len(i)):
        if header_list[j].strip() == '':
          print('here')
          if j == 0:
            text = text + ' ' + str(index_row) + '. ' + i[j].strip() + '\n'
          else:
            text = text + '    ' + i[j].strip() + '\n'
        else:
          if j == 0:
            text = text + ' ' + str(index_row) + '. ' + header_list[j].strip() + ': ' + i[j].strip() + '\n'
          else:
            text = text + '    ' + header_list[j].strip() + ': ' + i[j].strip() + '\n'
      table_text = table_text + text
  else:
    for i in row_list:
      text = ''
      for j in range(len(i)):
        if j == 0:
          text = text + ' > ' + i[j].strip() + '\n'
        else:
          text = text + '   ' + i[j].strip() + '\n'
      table_text = table_text + text
  return table_text
```
